```python
df = pd.read_csv(path + "/churn_Data.csv")
df
```

```python
# remove df["income"] max value
df = df[df["income"] != df["income"].max()]
```

```python
import matplotlib.pyplot as plt

# colored with churn
plt.scatter(df["income"], df["tenure"], c=df["churn"])
plt.title("Churn vs Income")
plt.xlabel("Gelir")
plt.ylabel("Tenure")
plt.show()
```

```python
# use knnto predict custcat
# use knnto predict custcat
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

x = df[df.columns[:-1]]
y = df["churn"]

x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=15, metric='minkowski')
knn.fit(x_train, y_train)
y = df["churn"]

x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=15, metric='minkowski')
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)

print("Accuracy:", metrics.accuracy_score(y_test, knn_pred))
```

```python
# confision matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,knn_pred)
```
